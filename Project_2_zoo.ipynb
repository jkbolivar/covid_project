```python
#Llamamos el set de datos público 
zoo_data = 'https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.data'

# los nombres de las columnas se encuentran en :  'https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.names'
zoo_name= ['animal name','hair','feathers','eggs','milk','airborne','aquatic','predator','toothed','backbone','breathes','venomous','fins','legs','tail','domestic',' cat-size','type']
```

```python
import pandas as pd
```

```python
#otra forma de instalar pandas
#!pip install pandas
```

```python
df_zoo=pd.read_csv(zoo_data)
```

```python
df_zoo.head(10)
```

```python
df_zoo.columns=zoo_name
```

```python
df_zoo.head()
```

```python
import altair as alt 
```

```python
plot_01=alt.Chart(df_zoo).mark_line().encode(
    x='animal name',
    y='legs'
)
```

```python
plot_01
```

```python
plot_02=alt.Chart(df_zoo).mark_bar().encode(
    x='animal name',
    y='fins'
)
```

```python
plot_02
```

```python
import plotly.express as px
```

```python
fig=px.pie(df_zoo,values='fins', names='animal name', title='Grafica de pie ', height=500)
fig.update_traces(textposition='inside')
fig.show()
```

```python
#mostrar una grafica de pie animales acuatico y no acuatico
df_acuatics = df_zoo[['animal name','aquatic']].groupby('aquatic').count().reset_index()
df_acuatics['aquatic'] = df_acuatics['aquatic'].apply(lambda x: 'Acuatico' if x ==0 else 'no acuatico')
df_acuatics.rename(columns ={'animal name':'count'}, inplace = True)
px.pie(df_acuatics, values = 'count',names='aquatic', title='Animales acuaticos')
```

```python
#mostrar grafica de pie en relación animales depredadores y no depredadores
df_acuatics = df_zoo[['animal name','aquatic']].groupby('aquatic').count().reset_index()
df_acuatics['aquatic'] = df_acuatics['aquatic'].apply(lambda x: 'Acuatico' if x ==0 else 'no acuatico')
df_acuatics.rename(columns ={'animal name':'count'}, inplace = True)
px.pie(df_acuatics, values = 'count',names='aquatic', title='Animales acuaticos')

df_predator = df_zoo[['animal name','predator']].groupby('predator').count().reset_index()
df_predator['predator'] = df_predator['predator'].apply(lambda x:'predator' if x==0 else 'no depredator')
df_predator.rename(columns= {'animal name':'count'}, inplace = True )
px.pie(df_predator, values ='count', names='predator', title='Animales depredadores')
```
